```python
import pandas as pd
import matplotlib.pyplot as plt
from pymongo import MongoClient
import datetime
import pprint
```

```python
recipe_df = pd.read_csv("data/RAW_recipes.csv")

recipe_df.head()
```

```python
recipe_df[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']] = recipe_df.nutrition.str.split(",", expand=True)

recipe_df['calories'] =  recipe_df['calories'].apply(lambda x: x.replace('[',''))

recipe_df['carbohydrates (PDV)'] =  recipe_df['carbohydrates (PDV)'].apply(lambda x: x.replace(']',''))

recipe_df[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']] = recipe_df[['calories','total fat (PDV)','sugar (PDV)','sodium (PDV)','protein (PDV)','saturated fat (PDV)','carbohydrates (PDV)']].astype('float')

recipe_df.head()
```

```python
recipe_df.drop(columns=['nutrition', 'contributor_id', 'submitted'], inplace=True)

recipe_df.head()

#recipe_df.to_csv("data/cleaned_recipes.csv")
```
